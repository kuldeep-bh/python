```python
x = np.linspace(0,2,100)
```

```python
plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'quadretic')
plt.plot(x, x**3, label = 'cubic')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('simple plot')
plt.legend()
plt.show()
```

above graph of 3 lines is state machine interface this can work with various mathematical functions
